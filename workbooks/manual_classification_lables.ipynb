```python
def contains_sub_text(df, column_name, sub_text, case=False):
    df = df[df[column_name].str.contains(sub_text, case=case)]
    print(f'df.shape: {df.shape}')
    print(f'df.{column_name}.nunique(): {df[column_name].nunique()}')
    display(df.head(5))
    display(df[column_name].value_counts().head(10))
    return df

def change_label(df, column_name, new_label, sub_text=False, case=False):
    if sub_text == False:
        sub_text = new_label
    df.loc[df[column_name].str.contains(sub_text, case=False), column_name] = new_label
    return df
```
