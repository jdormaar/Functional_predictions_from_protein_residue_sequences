```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
import json
```

```python
filename = '../data/pdb_seqres.txt'
```

```python
# opening the file
file_obj = open(filename, "r")
```

```python
# reading the data from the file
file_data = file_obj.read()
```

```python
# splitting the file data into lines
lines = file_data.splitlines()
# print(lines)
file_obj.close() 

lines[:10]
```

```python
pdb_mol = []
sequence = []
for line in lines:
    if line[:1]!='>':
        sequence.append(line)
    else:
        pdb_mol.append(line)
```

```python
print(len(sequence))
print(len(pdb_mol))
```
